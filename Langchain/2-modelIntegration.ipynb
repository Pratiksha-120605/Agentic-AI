### Streaming and batch

### Streaming

Definition:
Streaming is when the model returns output token by token (or chunk by chunk) as it generates, instead of waiting for the entire response to finish.

Key Points:

Output appears in real-time.

Useful for chat apps, web UIs, or long text generation.

Reduces apparent latency — the user sees partial results quickly.

Often used with chat models, LLMs, or generative APIs.

```python
model.invoke("write a 200 words paragrapgh on ai")
```

```python
for chunk in model.stream("Write a 200 words paragraph on ai"):
    print(chunk.text,end="|",flush=True)
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.content)
```

### Batch

Definition:
Batching is when you process multiple inputs at once in a single call to the model or embeddings function.

Key Points:

Efficient for bulk processing.

Reduces API calls and speeds up processing.

Model returns a list of outputs, one per input.

```python
responses=model.batch([
    "How agentic ai and ai agents are different from each others?",
    "How do airplanes fly?",
    "What is quantum computing?"
])
for response in responses:
    print(response)
```

```python
model.batch([
    "How agentic ai and ai agents are different from each others?",
    "How do airplanes fly?",
    "What is quantum computing?"
],
config={
    'max_concurrency':5, # limit to 5 parrel cells
}
)
```
